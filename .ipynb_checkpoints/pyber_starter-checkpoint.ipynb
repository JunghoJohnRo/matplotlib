```python
%matplotlib inline

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
pyber_df = pd.merge(ride_df, city_df, on="city")

# Display the data table for preview
pyber_df.head()
```

### Calculating Variables to plot Bubble Plot

```python
# Grouping Dataframe by Type and City
city_type_df = pyber_df.groupby(["type", "city"]) 

avg_fare = city_type_df["fare"].mean()
total_rides = city_type_df["ride_id"].count()
total_drivers = city_type_df["driver_count"].sum()/city_type_df["driver_count"].count()

bp_df = pd.DataFrame({"Average Fare ($) Per City":avg_fare.round(2), 
                      "Total Number of Rides Per City": total_rides, 
                      "Total Number of Drivers Per City": total_drivers})
bp_df.head()
```

### Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban = bp_df.loc["Urban",:]
suburban = bp_df.loc["Suburban",:]
rural = bp_df.loc["Rural",:]

ux = urban["Total Number of Rides Per City"]
sx = suburban["Total Number of Rides Per City"]
rx = rural["Total Number of Rides Per City"]

uy = urban["Average Fare ($) Per City"]
sy = suburban["Average Fare ($) Per City"]
ry = rural["Average Fare ($) Per City"]

uz = urban["Total Number of Drivers Per City"]
sz = suburban["Total Number of Drivers Per City"]
rz = rural["Total Number of Drivers Per City"]

# Build the scatter plots for each city types
# Incorporate the other graph properties
fig, ax = plt.subplots()
fig.suptitle("Pyber Ride Sharing Data (2016)", fontsize=13, fontweight="bold")

ax.scatter(ux, uy, s=uz*10, color="lightcoral", alpha=0.8, edgecolor="black", linewidth=1)
ax.scatter(sx, sy, s=sz*10, color="lightskyblue", alpha=0.8, edgecolor="black", linewidth=1)
ax.scatter(rx, ry, s=rz*10, color="gold", alpha=0.8, edgecolor="black", linewidth=1)

ax.set_xlabel("Total Number of Rides (Per City)")
ax.set_ylabel("Average Fare($)")
ax.grid()

# Create a legend (Googled how to create the same size for labels)
lgnd = plt.legend(loc="best", labels=["Urban", "Suburban", "Rural"], title="City Types")
for handle in lgnd.legendHandles:
    handle.set_sizes([60.0])
    
# Incorporate a text label regarding circle size (Googled how to place text outside the figure)
plt.text(42, 38, "Note:")
plt.text(42, 36.7, "Circle size correlates with driver count per city.")

# Save Figure (Had to google for how to get note to print with the figure)
plt.savefig("Bubble_Plot.png", bbox_inches='tight')

# Show plot
plt.show()
```

### Total Fares by City Type

```python
# Calculate Type Percents
fare_df = pyber_df.groupby("type")
total_fare = fare_df["fare"].sum()
total_fare_df = pd.DataFrame(total_fare).rename(columns={"fare":"Fares"})

# Build Pie Chart
fares = total_fare_df.keys()
total_fare_df.plot(kind="pie", y=fares, autopct="%1.1f%%", shadow=True, 
                   startangle=140, legend=False, explode=[0,0,0.1], 
                   colors=["gold", "lightskyblue", "lightcoral"], 
                   title=("% of Total Fares by City Type"))
plt.axis("equal")

# Save Figure

# Show Figure
plt.show()
```

### Total Rides by City Type

```python
# Calculate Ride Percents
rides_df = pyber_df.groupby("type")
totalrides = rides_df["ride_id"].count()
totalrides_df = pd.DataFrame(totalrides).rename(columns={"ride_id":"Rides"})

# Build Pie Chart
rides = totalrides_df.keys()
totalrides_df.plot(kind="pie", y=rides, autopct="%1.1f%%", shadow=True, 
                   startangle=140, legend=False, explode=[0,0,0.1], 
                   colors=["gold", "lightskyblue", "lightcoral"], 
                   title=("% of Total Rides by City Type"))
plt.axis("equal")

# Save Figure

# Show Figure
plt.show()
```

### Total Drivers by City Type

```python
# Calculate Driver Percents
drivers_df = bp_df["Total Number of Drivers Per City"]
totaldrivers = drivers_df.groupby("type").sum()
totaldrivers_df = pd.DataFrame(totaldrivers).rename(columns={"driver_count":"Drivers"})

# Build Pie Charts
drivers = totaldrivers_df.keys()
totaldrivers_df.plot(kind="pie", y=drivers, autopct="%1.1f%%", shadow=True, 
                     startangle=160, legend=False, explode=[0,0,0.1], 
                     colors=["gold", "lightskyblue", "lightcoral"], 
                     title=("% of Total Drivers by City Type"))
plt.axis("equal")

# Save Figure

# Show Figure
plt.show()
```
